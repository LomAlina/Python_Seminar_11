```python
import math
import numpy as np
```

```python
a, b, c = 5, 10, -30
```

```python
def func(x):
    function = a*x**2 + b*x + c
    return function
```

```python
def solution(a, b, c):
    d = b**2 - 4*a*c
    if d > 0:
        x1 = (-b + math.sqrt(d)) / (2 * a)
        x2 = (-b - math.sqrt(d)) / (2 * a)
        return (x1, x2)
    if d == 0:
        x = -b / (2 * a)
        return x
    else:
        print('Вещественных корней нет')
        return None
```

```python
x = np.arange(-10, 10, 0.1)
```

```python
extr_func = min(func(x))
```

```python
print(f'Корни функции {solution(a,b,c)}, экстремум функции {min(func(x))}')
```

```python
x = solution(a, b, c - extr_func)
```

```python
x1 = np.arange(-4, x, 0.1)
x2 = np.arange(x, 4, 0.1)
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title(f'Корни функции x1 = {round(solution(a,b,c)[0], 2)}, x2 = {round(solution(a,b,c)[1], 2)}')
plt.plot(x1, func(x1),'r', label='Убывание')
plt.plot(x2, func(x2),'b', label='Возрастание')
plt.plot(x, func(x), 'ro', label=f'Экстремум функции {extr_func}')
plt.text(x, func(x) + 300, f'Экстремум {extr_func}')
plt.legend()
plt.show()
```
